```python
import numpy as np

xTrainForSetsuta = np.array(irisDataset.data.features)

# Since we are dealing with strings as our target type, we must convert into integers, and binary.
# In particular Iris-setosa and Iris-versicolor will be identified as 0 and Iris-virginica will be 1.
# If the model predicts 1 we know it is an Iris-viginica, if it is a 0, we use a second model to conduct
# binary logistic regression again on the last two Iris species.

yTrainBeforeBinaryConversion = np.array(irisDataset.data.targets)

wForSetsuta = np.ones(4)

bForSetsuta = 1

wForVersicolor = np.ones(4)

bForVersicolor = 1

lambda_Setsuta = 0.1

lambda_Versicolor = 0.1

learningRateAlphaSetsuta = 0.01

learningRateAlphaVersicolor = 0.01

numIterations = 10000
print(xTrainForSetsuta)
```

```python
def computeGradients(xTrain, yTrain, w, b, lambda_):
    
    derivativeW = np.zeros(w.shape[0])
    derivativeB = 0
    
    for i in range(xTrain.shape[0]):
        z = (np.dot(xTrain[i],w) + b)
        f_x = (1 / (1 + np.exp(-1 * z)))
        f_x -= yTrain[i]
        for j in range(xTrain.shape[1]):
            derivativeW[j] = derivativeW[j] + (f_x * xTrain[i, j])
        derivativeB = derivativeB + f_x
    derivativeW = derivativeW/(xTrain.shape[1])
    derivativeB = derivativeB/(xTrain.shape[1])
    
    for i in range(xTrain.shape[1]):
        derivativeW[i] = (derivativeW[i] + ((lambda_/xTrain.shape[0]) * w[i]))
    return derivativeW, derivativeB
```

```python
def gradientDescent(xTrain, yTrain, w, b, lambda_, learningRateAlpha, numIterations):
    
    for i in range(numIterations):
        derivativeW, derivativeB = computeGradients(xTrain, yTrain, w, b, lambda_)
        w = (w - (learningRateAlpha * derivativeW))
        b = (b - (learningRateAlpha * derivativeB))
    return w, b
```

```python
# Train setsota model
wForSetsuta, bForSetsuta = gradientDescent(xTrainForSetsuta, yTrainForSetsuta, wForSetsuta, bForSetsuta, lambda_Setsuta, learningRateAlphaSetsuta, numIterations)

# Train versicolor model
wForVersicolor, bForVersicolor = gradientDescent(xTrainForVersicolor, yTrainForVersicolorBinary, wForVersicolor, bForVersicolor, lambda_Versicolor, learningRateAlphaVersicolor, numIterations)
```

```python
def irisDecision(wForSetsuta, bForSetsuta, wForVersicolor, bForVersicolor, XFeatures):
    
    decision = ""
    
    z = np.dot(XFeatures, wForSetsuta) + bForSetsuta
    f_x = (1 / (1 + np.exp(-1 * z)))
    if f_x >= 0.5:
        decision = "Iris-setsuta" + " precision: " + str(int(1000 * f_x)/10)
    if f_x < 0.5:
        z = np.dot(XFeatures, wForVersicolor) + bForVersicolor
        f_x = (1 / (1 + np.exp(-1 * z)))
        if f_x >= 0.5:
            decision = "Iris-versicolor" + " precision: " + str(int(1000 * f_x)/10)
        if f_x < 0.5:
            decision = "Iris-virginica" + " precision: " + str(int(1000 * (1 - f_x))/10)
    return decision
```
